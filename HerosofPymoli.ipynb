```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head(10)

```

## Player Count

* Display the total number of players


```python
unique_sn = purchase_data_df["SN"].nunique()


player_count_df = pd.DataFrame({"Total Players": [unique_sn]})
player_count_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data_df["Item ID"].nunique() 
average_price = purchase_data_df["Price"].mean() 
number_purchase = purchase_data_df["Price"].count() 
total_rev = purchase_data_df["Price"].sum()

purchase_df = pd.DataFrame({"Number of Unique Items":[unique_items], "Average Price":[average_price], "Number of Purchases":[number_purchase], "Total Revenue": [total_rev]})

purchase_df["Total Revenue"]= purchase_df["Total Revenue"].map("${:,.2f}".format)
purchase_df["Average Price"]= purchase_df["Average Price"].map("${:,.2f}".format)

purchase_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
uniqueplayer_df = purchase_data_df[["Gender","SN", "Age"]].drop_duplicates()


total_count = uniqueplayer_df["Gender"].count() 
gender_count = uniqueplayer_df["Gender"].value_counts()
#female_count = purchase_data_df["Gender" : "Female"].nunique()
#other_count = purchase_data_df["Gender" : "Other / Non-Disclosed"].nunique()

gender_percent = (gender_count / total_count)
#female_percent = female_count / total_count
#other_percent = other_count / total_count



gender_df = pd.DataFrame({"Total Count": gender_count, "Percentage of Players": gender_percent})
gender_df["Percentage of Players"]= gender_df["Percentage of Players"].map("{:,.2%}".format)
#gender_df

gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python



gender_count2 = purchase_data_df.groupby(["Gender"]).count()["Price"]

average_price = purchase_data_df.groupby(["Gender"]).mean()["Price"]

total_price = purchase_data_df.groupby(["Gender"]).sum()["Price"]

average_purchase_price = total_price / gender_count

puchase_table_df = pd.DataFrame({"Purchase Count":gender_count2, "Average Purchase Price":average_price, "Total Purchase Value": total_price, "Avg Total Purchase per Person": average_purchase_price})

puchase_table_df["Avg Total Purchase per Person"]= puchase_table_df["Avg Total Purchase per Person"].map("${:,.2f}".format)
puchase_table_df["Average Purchase Price"]= puchase_table_df["Average Purchase Price"].map("${:,.2f}".format)
puchase_table_df["Total Purchase Value"]= puchase_table_df["Total Purchase Value"].map("${:,.2f}".format)

puchase_table_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#list 1 - Nuimerical
age_bins = [0,9.999,14.999,19.999,24.999,29.999,34.999,39.999,999]

# List of Age names
age_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#slice the data
uniqueplayer_df["age_range"] = pd.cut(uniqueplayer_df["Age"],age_bins,labels=age_names) 

age_range_count = uniqueplayer_df["age_range"].value_counts()
percent_age_range = age_range_count / unique_sn

percent_age_range

#show DF

age_demo_df = pd.DataFrame({ "Total Count":age_range_count, "Percentage of Players":percent_age_range})


age_demo_df["Percentage of Players"]= age_demo_df["Percentage of Players"].map("{:,.2%}".format)


age_demo_df
age_demo_df.sort_index()
#age_range_count
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
totalpurchase_df = purchase_data_df
totalpurchase_df["age_range"] = pd.cut(purchase_data_df["Age"],age_bins,labels=age_names)


age_purchase_count = totalpurchase_df.groupby(["age_range"]).count()["Age"]

#average_price = purchase_data_df.groupby(["Gender"]).mean()["Price"]

#total_price = purchase_data_df.groupby(["Gender"]).sum()["Price"]

#average_purchase_price = total_price / gender_count




tot_purchase_value = totalpurchase_df.groupby(["age_range"]).sum()["Price"]
avrg_purch_price = tot_purchase_value / age_purchase_count


avrg_purch_pp = tot_purchase_value / age_range_count


age_purchase_table_df = pd.DataFrame({
                                "Age Range":age_names, 
                                "Purchase Count":age_purchase_count, 
                                "Average Purchase Price":avrg_purch_price,
                                "Total Purchase Value":tot_purchase_value,
                                "Avg Total Purchase per Person":avrg_purch_pp
                                })


age_purchase_table_df["Average Purchase Price"]= age_purchase_table_df["Average Purchase Price"].map("${:,.2f}".format)
age_purchase_table_df["Total Purchase Value"]= age_purchase_table_df["Total Purchase Value"].map("${:,.2f}".format)
age_purchase_table_df["Avg Total Purchase per Person"]= age_purchase_table_df["Avg Total Purchase per Person"].map("${:,.2f}".format)





age_purchase_table_df
age_purchase_table_df.sort_index()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Total_Purchase_Price = purchase_data_df.groupby(["Item ID", "Item Name"]).sum()['Price'].rename("Total Purchase Price")
Total_Purchase_Price
```

```python
Average_item_purchase = purchase_data_df.groupby(["Item ID", "Item Name"]).mean()['Price'].rename("Item Price")
Average_item_purchase
```

```python
Item_count = purchase_data_df.groupby(["Item ID", "Item Name"]).count()['Price'].rename("Purchase Count")
Item_count
```

```python


popular_item_df = pd.DataFrame({
                                "Purchase Count": Item_count, 
                                "Item Price":Average_item_purchase,
                                "Total Purchase Value": Total_Purchase_Price})


popular_item_df.sort_values("Purchase Count", ascending=False, inplace=True)

popular_item_df["Item Price"]= popular_item_df["Item Price"].map("${:,.2f}".format)
popular_item_df["Total Purchase Value"]= popular_item_df["Total Purchase Value"].map("${:,.2f}".format)


popular_item_df
```
